# outlier using inter quantile range

```python
q1,q3=np.percentile(x["humidity"],[25,75])
iqr=q3-q1
lower_bound=q1-(iqr*1.5)
upper_bound=q3+(iqr*1.5)
print(lower_bound,upper_bound)
x[(x["humidity"] <2.0) | (x["humidity"] >122.0)]
```

```python
x.loc[x["humidity"] <2.0,"humidity"]=2.0
#x.loc[x["humidity"] >122.0,"humidity"]=122.0
```

```python
q1,q3=np.percentile(x["windspeed"],[25,75])
iqr=q3-q1
lower_bound=q1-(iqr*1.5)
upper_bound=q3+(iqr*1.5)
print(lower_bound,upper_bound)
x[(x["windspeed"] <-7.993100000000002 ) | (x["windspeed"] > 31.992500000000003)]
```

```python
x.loc[x["windspeed"] < -7.993100000000002,"windspeed"]=-7.993100000000002
x.loc[x["windspeed"] > 31.992500000000003,"windspeed"]=31.992500000000003
```

```python
for i in x :
    sns.boxplot(y=i,data=x)
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lg= LinearRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
y_pred=lg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn import metrics 
```

```python
metrics.r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor 
  
 # create regressor object 
regressor = RandomForestRegressor(n_estimators = 100, random_state = 42) 
  
# fit the regressor with x and y data 
regressor.fit(X_train, y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
metrics.r2_score(y_test,y_pred)
```
