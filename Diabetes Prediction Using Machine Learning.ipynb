# checking for an outlier

```python
for i in X :
    sns.boxplot(y=i,data=X)
    plt.show()
```

```python
def impute_outlier(lebal) :
    q1,q3=np.percentile(X[lebal],[25,75])
    iqr=q3-q1
    lower_bound=q1-(iqr*1.5)
    upper_bound=q3+(iqr*1.5)
    return lower_bound,upper_bound
```

```python
l=[]
for i in X :
    l.append((impute_outlier(i)))
```

```python
l
```

```python
def update(lab,i) :
    print(i[0],i[1])
    X.loc[X[lab]<i[0],lab]=i[0]
    X.loc[X[lab]>i[1],lab]=i[1]
```

```python
X.columns
```

```python
update('Pregnancies',l[0])
update('Glucose',l[1])
update('BloodPressure',l[2])
update('SkinThickness',l[3])
update('Insulin',l[4])
update('DiabetesPedigreeFunction',l[5])
update('Age',l[6])
```

```python
for i in X :
    sns.boxplot(y=i,data=X)
    plt.show()
```

```python
for i in range(0,len(l)) :
    print(l[i])
```
